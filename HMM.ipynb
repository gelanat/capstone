### Sampling from an HMM
based on https://github.com/hmmlearn/hmmlearn

```python
import numpy as np
import matplotlib.pyplot as plt

from hmmlearn import hmm


#Building an HMM and setting parameters
model = hmm.GaussianHMM(n_components=4, covariance_type="full")

#Setting the estimated parameters
model.startprob_ = np.array([0.2, 0.5, 0.3, 0.0])
model.transmat_ = np.array([[0.7, 0.2, 0.0, 0.1],
                     [0.3, 0.5, 0.2, 0.0],
                     [0.0, 0.3, 0.5, 0.2],
                     [0.2, 0.0, 0.2, 0.6]])
model.means_ = np.array([[0.0,  0.0],
                  [0.0, 11.0],
                  [9.0, 10.0],
                  [11.0, -1.0]])
model.covars_ = .5 * np.tile(np.identity(2), (4, 1, 1))


#Generating samples
X, Z = model.sample(500)

#Plotting
plt.plot(X[:, 0], X[:, 1], ".-", label="observations", ms=6,
         mfc="orange", alpha=0.7)
for i, m in enumerate(model.means_):
    plt.text(m[0], m[1], 'Feature %i' % (i + 1),
             size=17, horizontalalignment='center',
             bbox=dict(alpha=.7, facecolor='w'), color='r')
plt.legend(loc='best')
plt.show()
```
